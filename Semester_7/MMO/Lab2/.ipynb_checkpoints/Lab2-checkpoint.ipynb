```python
class DataPipeline:

    def __init__(self):
        self.medians = None
        self.median_square = None
        self.median_floor = None
        self.median_price = None

    def fit(self, df):
        self.medians = df[['LifeSquare', 'Healthcare_1']].median()
        self.median_square = df['Square'].median()
        self.median_square = df['Floor'].median()
        self.median_square = df['Price'].median()

    def transform(self, df):

        # Выбросы
        square = 'Square'
        total_square_max = np.quantile(df[square], q=0.975)
        total_square_min = np.quantile(df[square], q=0.025)
        condition_square = (df[square] > total_square_max) | (df[square] < total_square_min)
        df.loc[condition_square, square] = self.median_square

        floor = 'Floor'
        total_floor_max = np.quantile(df[floor], q=0.975)
        total_floor_min = np.quantile(df[floor], q=0.025)
        condition_floor = (df[floor] > total_floor_max) | (df[floor] < total_floor_min)
        df.loc[condition_floor, floor] = self.median_floor

        price = 'Price'
        total_price_max = np.quantile(df[price], q=0.975)
        total_price_min = np.quantile(df[price], q=0.025)
        condition_price = (df[price] > total_price_max) | (df[price] < total_price_min)
        df.loc[condition_price, price] = self.median_price

        #Заполнение пропусков
        df[['LifeSquare', 'Healthcare_1']] = df[['LifeSquare', 'Healthcare_1']].fillna(self.medians)

        #Отношение жилой площади к общей
        df['Residential to gross floor area'] = df['LifeSquare'] / df['Square']
        #Привлекательность жилья в зависимости от экологии, года постройки и цены
        df['House attractivness'] = df['Ecology_1'] / max(df['Ecology_1']) + df['HouseYear'] / max(df['HouseYear']) - df['Price'] / max(df['Price'])

        return df

```

```python
tempData = data.copy(deep=True)
pipe = DataPipeline()
pipe.fit(tempData)
displayData = pipe.transform(tempData)
data.to_csv('./result.csv', index=False, encoding='utf-8')
displayData
```
